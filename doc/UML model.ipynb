```python




from graphviz import Source
temp = """
digraph G {
        fontname = "Sans"
        fontsize = 8
        compound=true
        splines=false

        node [
                fontname = "Sans"
                fontsize = 8
                shape = "record"
        ]

        edge [
                fontname = "Sans"
                fontsize = 8
        ]


        res [   style=filled, fillcolor=lightgrey,
                label = "{Relations to Resource|
                label: RelationToResource\l
                source: DNB, ZDB, SBB, KPE\l|
                + Source: string\l
                + SourceType: string\l
                + TypeAddInfo: string\l
                + TempValidity: string\l}"
        ]

        // Ressourcen
        Resource [
                label = "{Resource from Bibliographic File|
                label: Resource\l
                source: DNB, ZDB, SBB, KPE\l|
                + Id: string\l
                + Uri: string\l
                + Name: string\l
                + Creator: string\l
                + Date: string\l
                + DateApprox: string\l
                + Place: string\l
                + Lang: string\l
                + Genre: string\l
                + SourcePath: string\l}"
            ]
        
        rel2 [  style=filled, fillcolor=lightgrey,
                label = "{Relations to GND entities|
                labels: RelationToXXX\l
                source: DNB, ZDB, SBB, KPE\l|
                + Source: string\l
                + SourceType: string\l
                + TypeAddInfo: string\l
                + TempValidity: string\l}"
        ]
        

        // Attribute für Relationen zwischen GND-Entitäten
        rel [   style=filled, fillcolor=lightgrey,  
                label = "{Relations to GND entities|
                labels: RelationToXXX\l
                source: GND\l|
                + Source: string\l
                + SourceType: string\l
                + TypeAddInfo: string\l
                + TempValidity: string\l}"
        ]

        soc [   style=filled, fillcolor=lightgrey,
                label = "{Social Relation|
                label: SocialRelation\l
                source: computed\l|
                + Source: string\l
                + SourceType: string\l
                + TypeAddInfo: string\l}"
        ]
        
        chron [ style=filled, fillcolor=lightgrey,
                label = "{Relation to date|
                label: RelationToChronTerm\l
                source: GND\l|
                + Source: string\l
                + SourceType: string\l}"
        ]

        isil [  style=filled, fillcolor=lightgrey,
                label = "{Relation to ISIL body|
                label: RelationToIsilTerm\l
                source: GND\l|
                + TypeAddInfo: string\l}"
        ]
        
        // Zusätzliche Typen aus der GND

        IsilTerm [
                label = "{ISIL body|
                label: IsilTerm\l
                source: GND\l|
                + Id: string\l
                + Name: string\l}"
        ]
        
        ChronTerm [
                label = "{Date|
                label: ChronTerm\l
                source: GND\l|
                + Id: string\l
                + Name: string\l}"

        ]

        subgraph clusterAnimalImpl {
                label = "Original entities from Authority File (GND)"
                
//                attrib [
//                        label = "{general Attributes|
//                        + Id: string\l
//                        + OldId: string\l
//                        + Uri: string\l
//                        + GenType: string\l
//                        + SpecType: string\l
//                        + Name: string\l
//                        + VariantName: string\l}"
//                    ]
                

                TopicTerm [
                        label = "{Subject heading|
                        label: TopicTerm\l
                        source: GND\l|
                        + Id: string\l
                        + OldId: string\l
                        + Uri: string\l
                        + GenType: string\l
                        + SpecType: string\l
                        + Name: string\l
                        + VariantName: string\l|
                        + GenSubdiv: string\l
                        + Info: string\l}"
                ]

                UniTitle [
                        label = "{Work|
                         label: UniTitle\l
                        source: GND\l|
                        + Id: string\l
                        + OldId: string\l
                        + Uri: string\l
                        + GenType: string\l
                        + SpecType: string\l
                        + Name: string\l
                        + VariantName: string\l|
                        + Creator: string\l
                        + Date: string\l
                        + Medium: string\l
                        + Lang: string\l}"
                ]

                MeetName [
                        label = "{Conference or Event|
                        label: MeetName\l
                        source: GND\l|
                        + Id: string\l
                        + OldId: string\l
                        + Uri: string\l
                        + GenType: string\l
                        + SpecType: string\l
                        + Name: string\l
                        + VariantName: string\l|
                        + Place: string\l
                        + DateOriginal: string\l
                        + DateApproxBegin: string\l
                        + DateApproxEnd: string\l
                        + SubUnit: string\l
                        + Info: string\l}"
                ]

                GeoName [
                        label = "{Place or geographic name|
                        label: GeoName\l
                        source: GND\l|
                        + Id: string\l
                        + OldId: string\l
                        + Uri: string\l
                        + GenType: string\l
                        + SpecType: string\l
                        + Name: string\l
                        + VariantName: string\l|
                        + GenSubdiv: string\l
                        + Coordinates: string\l
                        + UriGeonames: string\l
                        + GeoArea: string\l
                        + Info: string\l}"
                ]

                CorpName [
                        label = "{Corporate Body|
                        label: CorpName\l
                        source: GND\l|
                        + Id: string\l
                        + OldId: string\l
                        + Uri: string\l
                        + GenType: string\l
                        + SpecType: string\l
                        + Name: string\l
                        + VariantName: string\l|
                        + DateOriginal: string\l
                        + DateApproxBegin: string\l
                        + DateApproxEnd: string\l
                        + DateStrictBegin: string\l
                        + DateStrictEnd: string\l
                        + SubUnit: string\l
                        + Info: string\l}"
                ]

                PerName [
                        label = "{Person|
                        label: PerName\l
                        source: GND\l|
                        + Id: string\l
                        + OldId: string\l
                        + Uri: string\l
                        + GenType: string\l
                        + SpecType: string\l
                        + Name: string\l
                        + VariantName: string\l|
                        + Gender: string\l
                        + DateOriginal: string\l
                        + DateApproxBegin: string\l
                        + DateApproxEnd: string\l
                        + DateStrictOriginal: string\l
                        + DateStrictBegin: string\l
                        + DateStrictEnd: string\l}"
                    ]

        }

        

        edge [
                arrowhead = "empty"
        ]


        edge [
                arrowhead = "none"

        ]

        GeoName -> rel [ltail=clusterAnimalImpl,]
        TopicTerm -> soc [ltail=clusterAnimalImpl,]
        MeetName -> chron [ltail=clusterAnimalImpl,]
        GeoName -> isil [ltail=clusterAnimalImpl,]
        Resource -> rel2
        Resource -> res
        
        edge [
               arrowhead = "normal"
//                headlabel = "0..*"
                label = "0..*"
        ]

        rel -> MeetName [lhead=clusterAnimalImpl]
        soc -> UniTitle [lhead=clusterAnimalImpl]
        chron -> ChronTerm
        isil -> IsilTerm
        rel2 -> PerName [lhead=clusterAnimalImpl]
        res -> Resource
}


"""
s = Source(temp, filename="UmlModel", format="svg")
# s.view()
s.render("UmlModel")
from IPython.display import SVG
SVG(filename='UmlModel.svg') 
```
